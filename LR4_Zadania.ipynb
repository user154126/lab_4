## 1. Создание Series из списка

Создайте Series из списка чисел [5, 15, 25, 35, 45] с индексами ['a', 'b', 'c', 'd', 'e'] . Выведите его на экран и определите его тип данных.

```python
import pandas as pd
s1 = pd.Series([5, 15, 25, 35, 45], ['a', 'b', 'c', 'd', 'e'])

print(s1)
```

## 2. Получение элемента Series

Дан Series с индексами ['A', 'B', 'C', 'D', 'E'] и значениями [12, 24, 36, 48, 60]. Используйте .loc[] для получения элемента с индексом 'C' и .iloc[] для получения третьего элемента.

```python
s1 = pd.Series([12, 24, 36, 48, 60], ['A', 'B', 'C', 'D', 'E'])

print(s1.loc['C'])
print(s1.iloc[2])
```

## 3. Фильтрация данных с помощью логической индексации.

Создайте Series из массива NumPy np.array([4, 9, 16, 25, 36, 49, 64]). Выберите только те элементы, которые больше 20, и выведите результат.

```python
import numpy as np
s1 = pd.Series(np.array([4, 9, 16, 25, 36, 49, 64]))
s1_20 = s1[s1 > 20]

print(s1_20)
```

## 5. Определение типа данных Series.

Создайте Series из списка ['cat', 'dog', 'rabbit', 'parrot', 'fish']. Определите тип данных с помощью .dtype , затем преобразуйте его в category с помощью .astype().

```python
s1 = pd.Series(['cat', 'dog', 'rabbit', 'parrot', 'fish'])

s1_dtype = s1.dtype

s1 = s1.astype('category')

s2_dtype = s1.dtype

print(s1_dtype)
print(s2_dtype)
```

## 6. Проверка пропущенных значений.

Создайте Series с данными [1.2, np.nan, 3.4, np.nan, 5.6, 6.8] . Напишите код, который проверяет, есть ли в Series пропущенные значения ( NaN ), и выведите индексы таких элементов.

```python
s1 = pd.Series([1.2, np.nan, 3.4, np.nan, 5.6, 6.8])

nan_index = s1[s1.isna()].index

print(nan_index)
```

## 7. Заполнение пропущенных значений.

Используйте Series из предыдущего задания и замените все NaN на среднее
значение всех непустых элементов. Выведите результат.

## 8. Арифметические операции с Series.

Создайте два Series:

1. s1 = pd.Series([10, 20, 30, 40], index=['a', 'b', 'c', 'd'])
2. s2 = pd.Series([5, 15, 25, 35], index=['b', 'c', 'd', 'e'])

Выполните сложение s1 + s2 . Объясните, почему в результате появляются NaN, и замените их на 0.

```python
s1 = pd.Series([10, 20, 30, 40], index=['a', 'b', 'c', 'd'])
s2 = pd.Series([5, 15, 25, 35], index=['b', 'c', 'd', 'e'])

result = s1 + s2
result_filled = result.fillna(0)

print(f"Результат сложения:\n{result}")
print()
print(f"Замена NaN на нули:\n{result_filled}")
```

Если индексы не совпадают, то на этих позициях появятся NaN. 

Индекс 'a' присутствует только в s1, а 'e' — только в s2, поэтому при сложении в этих местах будут NaN.

## 10. Основные статистические методы.

Создайте Series из 20 случайных чисел от 50 до 150 (используйте np.random.randint ). Найдите сумму, среднее, минимальное и максимальное значение. Выведите также стандартное отклонение.

```python
s1 = pd.Series(np.random.randint(50, 150, 20))

summ_s1 = s1.sum()
mean_s1 = s1.mean()
min_s1 = s1.min()
max_s1 = s1.max()
std_s1 = s1.std()

print(f"Сумма: {summ_s1}")
print(f"Среднее: {mean_s1}")
print(f"Минимальное значение: {min_s1}")
print(f"Максимальное значение: {max_s1}")
print(f"Стандартное отклонение: {std_s1}")
```

## 12. Проверка уникальности индексов.

Создайте Series с индексами ['A', 'B', 'A', 'C', 'D', 'B'] и значениями [10, 20, 30, 40, 50, 60] . Проверьте, являются ли индексы уникальными. Если нет, сгруппируйте повторяющиеся индексы и сложите их значения.

```python
s1 = pd.Series([10, 20, 30, 40, 50, 60], ['A', 'B', 'A', 'C', 'D', 'B'])

is_unique = s1.index.is_unique
unique_group = s1.groupby(s1.index).sum()

print(is_unique)
print()
print(unique_group)
```

## 15. Построение графика на основе Series.

Создайте Series , где индексами будут даты с 1 по 30 марта 2024 года, а
значениями – случайные числа от 50 до 150. Постройте график значений с
помощью matplotlib. Добавьте заголовок, подписи осей и сетку.

```python
import matplotlib.pyplot as plt
s1 = pd.Series(np.random.randint(50, 151, 30), pd.date_range(start="2024-03-01", periods=30, freq="D"))

plt.figure(figsize=(10, 8))
plt.plot(s1.index, s1.values, marker="o", linestyle="--", color="grey", label="Цена")

plt.title("Построение графика на основе Series")
plt.xlabel("Дата")
plt.ylabel("Цена")
plt.grid(True)
plt.legend()

plt.show()
```

## Вариант №13. Отслеживание колебаний акций.

Создайте Series , где индексами будут даты с 1 по 20 июля 2024 года, а значениями – случайные цены акции (от 90 до 110). Найдите дни с изменением цены более чем на 3% по сравнению с предыдущим
днём, выделите их на графике маркерами.

```python
dates = pd.date_range('2024-07-01', periods=20, freq='D')
prices = pd.Series(np.random.uniform(90, 110, 20), index=dates)

price_change = prices.pct_change() * 100
change_more_3 = price_change[abs(price_change) > 3]

plt.figure(figsize=(10, 6))
plt.plot(prices.index, prices.values, marker="o", linestyle="-", color="blue", label="Цена акции")
plt.scatter(change_more_3.index, prices[change_more_3.index], color="red", label="Изменение > 3%", zorder=2)

plt.title("Колебания цены акций c 1 по 20 июля 2024")
plt.xlabel("Дата")
plt.ylabel("Цена")
plt.grid()
plt.legend()
plt.show()
```
